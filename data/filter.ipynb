```python
# read the filtered_complaints.csv file and remove the narrative column

import pandas as pd

df = pd.read_csv('./filtered_complaints.csv')
print(df.shape)
df.head()
# df.drop(columns=['Consumer complaint narrative'], inplace=True)
# then save the df to a file to modify the old file

# df.to_csv('./filtered_complaints.csv', index=False)
```

```python
# Stratified sampling by product (30% per category)
df_sampled = (
    df.groupby('Product', group_keys=False)  # Avoid adding group key to index
    .apply(lambda x: x.sample(frac=0.1, random_state=42))  # Fixed seed for reproducibility
    .reset_index(drop=True)  # Clean index
)

# Verify distribution
print(df_sampled['Product'].value_counts(normalize=True))
```
